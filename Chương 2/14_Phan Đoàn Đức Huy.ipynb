```python
import sqlite3
import pandas as pd

# Tạo kết nối cơ sở dữ liệu SQLite trong bộ nhớ
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Khởi tạo bảng và chèn dữ liệu
cursor.executescript('''
CREATE TABLE course (
    id INTEGER PRIMARY KEY,
    course_name TEXT
);

CREATE TABLE student (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
);

INSERT INTO course (id, course_name) VALUES
(12, 'Giai tich'),
(34, 'Thong ke'),
(26, 'Tin hoc');

INSERT INTO student (student_id, name, class, course_id, score) VALUES
(1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
(2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
(3, 'Pham Van Nam', 'Toan Tin', NULL, 7.9),
(4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
(5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
(6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
(7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
(8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
(9, 'Duong Huu Phuc', 'Kinh Te', NULL, 7.2),
(10, 'Cao Thi Hanh', 'May Tinh', NULL, 7.0);
''')
conn.commit()

# Kiểm tra dữ liệu
display(pd.read_sql_query('SELECT * FROM student;', conn))
display(pd.read_sql_query('SELECT * FROM course;', conn))
```

# Bài 1

```python
# 1. Tích Descartes
df_cartesian = pd.read_sql_query('SELECT * FROM student, course;', conn)
print("1. Tích Descartes:")
display(df_cartesian)

# 2. INNER JOIN
df_inner = pd.read_sql_query('SELECT * FROM student INNER JOIN course ON student.course_id = course.id;', conn)
print("2. INNER JOIN:")
display(df_inner)

# 3. LEFT JOIN
df_left = pd.read_sql_query('SELECT * FROM student LEFT JOIN course ON student.course_id = course.id;', conn)
print("3. LEFT JOIN:")
display(df_left)
# 4 RIGHT
try:
    df_right = pd.read_sql_query('SELECT * FROM student RIGHT JOIN course ON student.course_id = course.id;', conn)
    print("4. RIGHT JOIN:")
    display(df_right)
except Exception as e:
    print("4. RIGHT JOIN bị lỗi do giới hạn của SQLite:", e)
#5 FULL OUTER JOIN
try:
    df_full = pd.read_sql_query('SELECT * FROM student FULL OUTER JOIN course ON student.course_id = course.id;', conn)
    print("5. FULL OUTER JOIN:")
    display(df_full)
except Exception as e:
    print("5. FULL OUTER JOIN bị lỗi do giới hạn của SQLite:", e)
```

# KẾT LUẬN CÂU 1:

+ Tích Descartes (CROSS JOIN):

Kết quả: Trả về 30 bản ghi (10 sinh viên x 3 môn học).

Kết luận: Phép toán này ghép mỗi dòng của bảng student với tất cả các dòng của bảng course mà không cần điều kiện. Kết quả này mang tính chất tổ hợp, thường chứa nhiều dữ liệu không có ý nghĩa thực tế (ví dụ: ghép sinh viên đang học Toán với thông tin môn Văn).

+ INNER JOIN:

Kết quả: Chỉ trả về 3 bản ghi (Sinh viên Nguyễn Minh Hoàng khớp mã 12; Trần Thị Lan và Bùi Tiến Dũng khớp mã 34).

Kết luận: INNER JOIN chỉ giữ lại các giao điểm hợp lệ. Những sinh viên chưa đăng ký môn (course_id = NULL) hoặc đăng ký mã môn không tồn tại trong bảng course (mã 20, 24) sẽ bị loại bỏ. Ngược lại, môn "Tin học" (mã 26) chưa có ai đăng ký cũng không được hiển thị.

+ LEFT JOIN:

Kết quả: Trả về toàn bộ 10 bản ghi của bảng student (bảng bên trái).

Kết luận: Phép nối này ưu tiên bảng sinh viên. 3 sinh viên có mã môn hợp lệ sẽ hiển thị kèm tên môn học. 7 sinh viên còn lại (mã môn NULL, 20, 24) vẫn được giữ lại trong danh sách nhưng các cột thông tin về môn học (id, course_name) sẽ bị để trống (NaN/None).

+ RIGHT JOIN:

Kết quả: Trả về thông tin của những sinh viên khớp mã môn, cộng thêm dòng của môn "Tin học" (mã 26).

Kết luận: Ưu tiên bảng course (bảng bên phải). Môn "Tin học" (mã 26) dù chưa có sinh viên nào đăng ký vẫn được giữ lại, các cột thông tin sinh viên tương ứng với môn này sẽ bị bỏ trống (NaN/None). Sinh viên đăng ký mã môn sai hoặc NULL bị loại bỏ.

+ FULL OUTER JOIN:

Kết quả: Là sự kết hợp của LEFT JOIN và RIGHT JOIN.

Kết luận: Giữ lại toàn bộ 10 sinh viên (dù đăng ký sai mã hoặc chưa đăng ký) và giữ lại toàn bộ 3 môn học (kể cả môn Tin học chưa có ai học). Các ô không có dữ liệu tương ứng sẽ được điền khoảng trống (NaN/None).

```python
# 2.1 Cập nhật dữ liệu thiếu và loại bỏ dữ liệu sai
cursor.executescript('''
-- Điền mã môn học tồn tại trong bảng course (VD: 26 - Tin học) cho các bạn thiếu
UPDATE student SET course_id = 26 WHERE course_id IS NULL;

-- Xóa các bạn có course_id không nằm trong bảng course
DELETE FROM student WHERE course_id NOT IN (SELECT id FROM course);
''')
conn.commit()

# a. Thống kê theo lớp
df_2a = pd.read_sql_query('''
    SELECT class, COUNT(student_id) AS tong_sv, AVG(score) AS diem_tb
    FROM student GROUP BY class;
''', conn)
print("a. Thống kê theo lớp:")
display(df_2a)

# b. Thống kê theo môn học
df_2b = pd.read_sql_query('''
    SELECT course_id, COUNT(student_id) AS tong_sv, AVG(score) AS diem_tb
    FROM student GROUP BY course_id;
''', conn)
print("b. Thống kê theo môn học:")
display(df_2b)

# c. Phân loại thi đua
df_2c = pd.read_sql_query('''
    SELECT course_id, AVG(score) AS diem_tb,
        CASE 
            WHEN AVG(score) >= 9.0 THEN 'Xuất sắc'
            WHEN AVG(score) >= 6.0 AND AVG(score) <= 8.9 THEN 'Tốt'
            WHEN AVG(score) < 6.0 THEN 'Kém'
        END AS phan_loai
    FROM student GROUP BY course_id;
''', conn)
print("c. Phân loại thi đua theo môn:")
display(df_2c)
```

# BÀI 3

```python
# Xếp hạng sinh viên
query_3 = '''
WITH Ranked AS (
    SELECT student_id, name, class, course_id, score,
           RANK() OVER(ORDER BY score DESC) as rank_chung,
           RANK() OVER(PARTITION BY class ORDER BY score DESC) as rank_lop,
           RANK() OVER(PARTITION BY course_id ORDER BY score DESC) as rank_mon,
           RANK() OVER(ORDER BY score ASC) as rank_chung_thap
    FROM student
)
SELECT * FROM Ranked;
'''
df_3 = pd.read_sql_query(query_3, conn)
print("Bảng xếp hạng sinh viên:")
display(df_3)

print("Top 3 sinh viên điểm cao nhất (Chung):")
display(df_3[df_3['rank_chung'] <= 3][['student_id', 'name', 'score', 'rank_chung']])

print("Top 3 sinh viên điểm thấp nhất (Chung):")
display(df_3[df_3['rank_chung_thap'] <= 3][['student_id', 'name', 'score', 'rank_chung_thap']])
```

```python
# Thêm trường graduation_date và tính toán
cursor.executescript('''
ALTER TABLE student ADD COLUMN graduation_date DATETIME;

WITH RankedStudents AS (
    SELECT student_id, RANK() OVER(ORDER BY score DESC) AS rank_val
    FROM student
)
UPDATE student
SET graduation_date = DATE('now', '+' || (
    SELECT rank_val FROM RankedStudents WHERE RankedStudents.student_id = student.student_id
) || ' days');
''')
conn.commit()

# Kiểm tra kết quả
df_4 = pd.read_sql_query('SELECT * FROM student;', conn)
print("Bảng sinh viên sau khi tính thời gian tốt nghiệp:")
display(df_4)
```

# KẾT LUẬN CÂU 4: Cập nhật Thời gian Tốt nghiệp


+ Về thao tác: Câu lệnh ALTER TABLE ... ADD COLUMN đã bổ sung thành công trường dữ liệu graduation_date với định dạng thời gian tiêu chuẩn (DATETIME) để lưu trữ chính xác thông tin ngày giờ.


+ Về kết quả tính toán: Hàm thời gian (về mặt logic học thuật là DATE_ADD ) đã tịnh tiến thành công mốc thời gian hiện tại dựa trên số hạng tương ứng của từng sinh viên.

+ Nhận xét dữ liệu: Kết quả cho thấy sự phân bổ ngày tốt nghiệp hoàn toàn khớp với năng lực học tập. Những bạn xuất sắc đạt thứ hạng 1 (Trần Thị Lan, Bùi Tiến Dũng) có thời gian tốt nghiệp sớm nhất (chỉ cộng thêm 1 ngày), trong khi bạn xếp hạng 6 (Nguyễn Minh Hoàng) sẽ có thời gian tốt nghiệp muộn nhất do bị cộng dồn thêm tới 6 ngày.
